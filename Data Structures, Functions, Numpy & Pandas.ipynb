```python
numbers = [10, 20, 30, 40, 50]#creation of List
print("Original list:", numbers)

numbers.append(60)#here add 60 to list
print("After append:", numbers)

numbers.extend([70, 80])#here add 70 and 80 to list
print("After extend:", numbers)

numbers.insert(2, 25)#here we are inserting an element first is the index position and second is the number we should insert 
print("After insert:", numbers)

numbers.remove(30)#the number 30 is remove form list
print("After remove:", numbers)

numbers.pop()#the pop remove the last number form list
print("After pop:", numbers)

print("Index of 40:", numbers.index(40))#index number of 40

print("Count of 20:", numbers.count(20))#the apperences of that number

numbers.sort()
print("After sort:", numbers)#sort in ascending order


numbers.reverse()
print("After reverse:", numbers)#the sorted number will be show in descending order

numbers.clear()
print("After clear:", numbers)#the list wil complete clear
```

```python
marks = int(input("Enter your marks: "))#getting input
if marks >= 90: #Condition statement: the mark should be greater than or equal to 90 
   print("Grade: A")
elif marks >= 75 and marks <= 89:#the mark should be greater than 75 and less than 89 
   print("Grade: B")
elif marks >= 50 and marks <= 74:#mark should be greater than 50 and less than 74
    print("Grade: C")
else:
    print("Fail")#if is not  come under any condition so is considered as Fail
```

```python
num = int(input("Enter a number: "))#getting an input
if num > 0:   #checking if the number is positive
   print("The number is positive")
   if num % 2 == 0:#If it is postive next it should be an even number
      print("It is also even")
   else:
      print("It is positive but odd")#else it is postive and odd number
elif num == 0: #if it is zero 
    print("The number is zero")

else:  #next if the number is not greater than zero is considered as a negative number
    print("The number is negative")
```

```python
import pandas as pd

data = {
    'Name': ['Arun', 'Meena', 'Ravi', 'Sita', 'Kumar', 'Anita', 'Ravi', 'John', 'Priya', 'Arun'],
    'Age': [23, 25, 30, 22, 28, 26, 30, 24, 27, 23],
    'City': ['Chennai', 'Coimbatore', 'Madurai', 'Salem', 'Erode', 'Trichy', 'Madurai', 'Chennai', 'Erode', 'Chennai'],
    'Salary': [30000, 35000, None, 28000, 40000, None, 32000, 29000, 36000, 30000]
} #creation and columns: 'Name', 'Age', 'City', and 'Salary'. 

df = pd.DataFrame(data)

print("Info:")
print(df.info())#info of data 

print("\nDescribe:")
print(df.describe())#describe data count, mean, max, std

print("\nSelected Columns:")
print(df[['Name', 'City']])#selectind only name and city columns only

df = df.drop('City', axis=1)
print("\nAfter Dropping City Column:")#Droping the city columns 
print(df)

df['Salary'].fillna(df['Salary'].mean(), inplace=True)
print("\nAfter Filling Missing Salary:")
print(df)

df = df.drop_duplicates()
print("\nAfter Removing Duplicates:")#Remove the duplicate 
print(df)


```
