```python
num = int(input("Enter a number: "))
if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print(num, "is not a prime number")
            break
        else:
            print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

```python
import random
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)
answer = int(input(f"What is the product of {num1} and {num2}? "))
if answer == num1 * num2:
    print("Correct! Well done.")
else:
    print("Wrong answer.")
    print("The correct answer is:", num1 * num2)
```

```python
room_charge_per_day = 2000
days_stayed = 4
doctor_fee = 1500
lab_test_cost = 300
number_of_tests = 3
medicine_charges = 2400
room_total = room_charge_per_day * days_stayed
lab_total = lab_test_cost * number_of_tests
total_bill = room_total + doctor_fee + lab_total + medicine_charges
if total_bill > 10000:
    discount = total_bill * 0.10
else:
    discount = 0

final_bill = total_bill - discount

print("Room Charges:", room_total)
print("Doctor Fee:", doctor_fee)
print("Lab Test Charges:", lab_total)
print("Medicine Charges:", medicine_charges)
print("Total Bill before Discount:", total_bill)
print("Discount:", discount)
print("Final Bill:", final_bill)

```
